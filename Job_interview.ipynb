```python
import pandas as pd

# Load CSV file
# Try this first
df = pd.read_csv("Software Questions.csv", encoding='latin1')


# Show sample rows
print(df.head())
```

```python
# Remove duplicates
df = df.drop_duplicates()
```

```python
print("Unique Topics:", df['Category'].unique())
```

```python
# Create a list of question dictionaries
question_bank = []

for index, row in df.iterrows():
    question_data = {
        "Category": row['Category'],
        "Question": row['Question'],
        "Difficulty": row.get('Difficulty', 'medium')  # use 'medium' as default
    }
    question_bank.append(question_data)
```

```python
for q in question_bank:
    print(f"Category: {q['Category']}")
    print(f"Question: {q['Question']}")
    print(f"Difficulty: {q['Difficulty']}")
    print("-" * 40)
```

```python
import json

# Save as a structured JSON file
with open("question_bank.json", "w", encoding='utf-8') as f:
    json.dump(question_bank, f, indent=4)
```

```python
from collections import defaultdict

grouped_questions = defaultdict(list)

for q in question_bank:
    grouped_questions[q["Category"]].append({
        "Question": q["Question"],
        "Difficulty": q["Difficulty"]
    })

# Save grouped by category
with open("grouped_questions.json", "w", encoding='utf-8') as f:
    json.dump(grouped_questions, f, indent=4)
```

```python
import requests

TOGETHER_API_URL = "https://api.together.xyz/v1/completions"
TOGETHER_API_KEY = "506df118097ac6706a6744823e2bc4b69a525f48172755d4cdbf455849140f17"

headers = {
    "Authorization": f"Bearer {TOGETHER_API_KEY}",
    "Content-Type": "application/json"
}

def evaluate_answer_together(question, correct_answer, user_answer):
    prompt = f"""
You are an expert tutor evaluating a student's answer.

Question:
{question}

Correct answer:
{correct_answer}

Student's answer:
{user_answer}

Please provide a detailed evaluation in the following format.
Speak directly to the student in second person ("you", "your").

1. **Step-by-step Reasoning**:
   Compare the student's answer to the correct answer. Explain what parts of the answer are right or wrong, point by point, as if you're speaking directly to the student. Use phrases like "You correctly explained...", "You missed mentioning...", "You misunderstood...".

2. **Constructive Feedback**:
   Offer guidance in second person: tell the student what they did well and how they can improve. Use phrases like "You should...", "Try to...", "Consider explaining...".

3. **Score**:
   Give a score from 0 to 5 using the rubric below.

5 - Excellent: Completely correct, thorough, clear reasoning, no errors.
4 - Good: Mostly correct with minor errors, good reasoning.
3 - Average: Partially correct, some key points missing, some errors.
2 - Poor: Mostly incorrect, flawed reasoning.
1 - Very Poor: Irrelevant or nonsensical, no understanding,random letters, usage of phrases like "i dont know","I am not sure","i dont remember".
0 - No attempt: Blank or no answer.

Only output the above format.
"""

    data = {
        "model": "mistralai/Mistral-7B-Instruct-v0.3",
        "prompt": prompt,
        "max_tokens": 700,
        "temperature": 0.7,
        "top_p": 1,
        # "stop": ["\n\n"]  # Commented out because it may prematurely stop output
    }

    response = requests.post(TOGETHER_API_URL, headers=headers, json=data)

    if response.status_code == 200:
        resp_json = response.json()
        try:
            # Together API returns completion text inside 'choices' list
            return resp_json['choices'][0]['text'].strip()
        except (KeyError, IndexError):
            return "Error: Unexpected response format."
    else:
        return f"Error: {response.status_code} - {response.text}"
```

```python
import random
import re  # for score parsing

topic = input("Choose a topic: ").strip()
num_questions = int(input("How many questions do you want? "))

questions = question_data.get(topic, [])

if not questions:
    print("Invalid topic or no questions available.")
else:
    if num_questions > len(questions):
        print(f"Only {len(questions)} questions available. Reducing number.")
        num_questions = len(questions)

    used_indices = set()
    total_score = 0.0  # float for decimals

    for i in range(num_questions):
        while True:
            idx = random.randint(0, len(questions) - 1)
            if idx not in used_indices:
                used_indices.add(idx)
                break

        q = questions[idx]
        question_text = q['Question']
        correct_answer = q.get('Answer', 'Not available')

        print(f"\nQuestion {i+1}: {question_text}")
        answer = input("Your Answer: ")

        print("\n--- Evaluation ---")
        result = evaluate_answer_together(question_text, correct_answer, answer)
        print(result)
        print("------------------")

        # Robust score extraction:
        score = 0.0
        try:
            lines = result.splitlines()
            score_line_index = None
            for idx_line, line in enumerate(lines):
                if "score" in line.lower():
                    score_line_index = idx_line
                    break

            if score_line_index is not None:
                # Check lines after the score line for numbers 0 to 5 (including decimals)
                found_score = False
                # Search next 3 lines after "Score" for number
                for check_line in lines[score_line_index + 1: score_line_index + 4]:
                    match = re.search(r"\b([0-5](?:\.\d+)?)\b", check_line)
                    if match:
                        score = float(match.group(1))
                        found_score = True
                        break

                if not found_score:
                    # As fallback, also check the score line itself for any number 0-5
                    match = re.search(r"\b([0-5](?:\.\d+)?)\b", lines[score_line_index])
                    if match:
                        score = float(match.group(1))
                    else:
                        print(" Could not find a valid score number.")
            else:
                print(" Could not find 'score' line in evaluation output.")

            print(f" Score for Question {i+1}: {score}/5")
        except Exception as e:
            print(" Could not parse score for this question.", e)

        total_score += score

    print(f"\n Your Total Score: {total_score} / {5 * num_questions}")
```
